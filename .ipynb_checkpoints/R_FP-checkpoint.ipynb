```python
import pandas as pd
import numpy as np
import torch 
import torch.nn as nn
import torch.optim as optim 

torch.manual_seed(12)
r_f=pd.read_csv("train.csv")

#normalizing the data 
def norm(x,col):
    mean_value=x[col].mean()
    std_value=x[col].std()
    x[col]=(x[col]-mean_value)/std_value
    return x[col]
r_f1=norm(r_f,'pressure')
r_f2=norm(r_f,'maxtemp')
r_f3=norm(r_f,'temparature')
r_f4=norm(r_f,'mintemp')
r_f5=norm(r_f,'dewpoint')
r_f6=norm(r_f,'humidity')
r_f7=norm(r_f,'cloud')
r_f8=norm(r_f,'sunshine')
r_f9=norm(r_f,'winddirection')
r_f10=norm(r_f,'windspeed')

#dropping the id column
r_f=r_f.drop(['id'],axis=1)


#extracting the true labels from the data ( i.e the rainfall column)
y_true=r_f.rainfall
y_true=torch.tensor(y_true,dtype=torch.float32)

#extracting the columns for the input data processing 
x_data=r_f.iloc[:,1:-1].values
x_data=torch.tensor(x_data,dtype=torch.float32)




#Neural net model 
class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.hidden_layers=nn.Sequential(
        nn.Linear(10,80),nn.ReLU(),
        nn.Linear(80,80),nn.Tanh(),
        nn.Linear(80,80),nn.ReLU(),
        nn.Linear(80,1),nn.Sigmoid()
        )

    def forward(self,x):
      y=self.hidden_layers(x)
      return y


        
from sklearn.metrics import roc_auc_score

class AUCLoss(nn.Module):
    def forward(self, y_pred, y_true):
        """
        A differentiable approximation of the AUC loss.
        y_pred: Model outputs (logits before sigmoid)
        y_true: True binary labels (0 or 1)
        """
        y_true = y_true.float()

        # Compute AUC (detach to prevent computation graph issues)
        auc_score = roc_auc_score(y_true.cpu().detach().numpy(), y_pred.cpu().detach().numpy())

        # Residual loss: 1 - AUC (to minimize)
        auc_loss = 1.0 - auc_score

        return torch.tensor(auc_loss, requires_grad=True)

# Define BCE loss
bce_loss = nn.BCEWithLogitsLoss()
auc_loss = AUCLoss()

# Model initialization
R_F= Model()
optimizer = optim.AdamW(R_F.parameters(), lr=0.00005)

# Hyperparameters for weighting the losses
lambda1 = 1  # Weight for BCE loss
lambda2 = 3  # Weight for AUC loss

#graph plotting terms
cost_items=[]
no_iters=[]

#smooth factor is used so that the updates of the weights happen smoothly.. ( but skip this process if needed)
smooth_factor = 0.5
y_true_smooth = y_true * (1 - smooth_factor) + smooth_factor / 2



# Training loop
epochs = 2500
epsilon= 1e-6
for epoch in range(epochs):
    optimizer.zero_grad()
    
    y_hat = R_F(x_data).squeeze()  # Model predictions (logits)
    #y_hat = torch.round(y_hat * 10) / 10      
    # Compute BCE loss
    loss_bce = bce_loss(y_hat, y_true_smooth)
    
    # Compute AUC loss
    loss_auc = auc_loss(y_hat, y_true)

    # Dynamically adjust weights (you can skip this step if needed)
    lambda1 = 1 / (loss_bce.item() + epsilon)
    lambda2 = 2 / (loss_auc.item() + epsilon)

    # Normalize so that lambda1 + lambda2 = 1
    lambda1 = lambda1 / (lambda1 + lambda2)
    lambda2 = lambda2 / (lambda1 + lambda2)

    # Weighted combination of losses
    total_loss = lambda1 * loss_bce + lambda2 * loss_auc
    cost_items.append(total_loss.detach())
    no_iters.append(epoch)
    total_loss.backward()
    optimizer.step()

    if epoch % 250 == 0:
        print(f"Epoch {epoch}, BCE Loss: {loss_bce.item():.4f}, AUC Loss: {loss_auc.item():.4f}, Total Loss: {total_loss.item():.4f}")
```

```python
import matplotlib.pyplot as plt



cost_items=np.array(cost_items)
plt.plot(no_iters[:],cost_items[:],linestyle='dashdot')
plt.title('Cost vs Epoch')
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.grid()
plt.show()
```
